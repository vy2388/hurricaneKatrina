```python
grouped_cols = ['Name','storm_name','year','Begin Date','End Date','Total CPI-Adjusted Cost (Millions of Dollars)','Deaths']
agg_columns = [col for col in df_cleaned.columns if col not in grouped_cols]
agg_functions = ['min','max','mean','median']
non_num = df_cleaned[agg_columns].select_dtypes(exclude='number').columns
agg_columns = [col for col in agg_columns if col not in non_num]

agg_dict = {col: agg_functions for col in agg_columns}
grouped_df = df_cleaned.groupby(grouped_cols).agg(agg_dict).reset_index()
```
