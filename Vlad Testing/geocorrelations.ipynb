```python
from haversine import haversine
from datetime import datetime

def calculate_temporal_difference(date1, date2):
    """Calculate the absolute difference in days between two dates."""
    return abs((date2 - date1).days)

def find_nearby_point(lat, lon, date, lat_lon_date_df, spatial_threshold, temporal_threshold, increments):
    """
    Find a point in lat_lon_date_df that is within a threshold distance from (lat, lon) and
    within a threshold temporal difference from date.
    
    Parameters:
        lat, lon: Coordinates of the reference point.
        date: Datetime object of the reference point.
        lat_lon_date_df: DataFrame containing latitude, longitude, and datetime columns.
        spatial_threshold: Distance threshold for considering points as nearby.
        temporal_threshold: Temporal difference threshold for considering points as nearby.
        increments: Increment value to increase the search radius.
        
    Returns:
        Information about the nearby point in lat_lon_date_df, or None if no such point exists.
    """
    i = 0
    while i <= spatial_threshold:
        for idx, row in lat_lon_date_df.iterrows():
            # Spatial proximity check
            if haversine((lat, lon), (row['latitude'], row['longitude'])) <= i:
                # Temporal proximity check
                if calculate_temporal_difference(date, row['datetime']) <= temporal_threshold:
                    # return f"index: {idx} | Distance (km): {i} | Date: {row['datetime']}, EventID: {row['EVENT_ID']}"
                    return [idx, i, row['datetime'], row['EVENT_ID']]
        i += increments
    return None
```

```python
import pandas as pd
import requests
from tqdm import tqdm
import json

def save_response_to_file(response, file_name, save_dir):
    with open(f"{save_dir}/{file_name}", 'w', encoding='utf-8') as f:
        json.dump(response, f, ensure_ascii=False, indent=4)
    return




def get_weatherJsonResponses(df, api_key, save_dir):
    for i, row in tqdm(df.iterrows(), total=df.shape[0]):
        latitude = row['lat']
        longitude = row['lon']
        date = row['date']
        # time = row['time_only']
        file_name = f"{row['ID']}.json"
        
        url = f"https://customer-archive-api.open-meteo.com/v1/archive?latitude={latitude}&longitude={longitude}&start_date={date}&end_date={date}&hourly=temperature_2m,relativehumidity_2m,dewpoint_2m,apparent_temperature,precipitation,rain,snowfall,snow_depth,weathercode,pressure_msl,surface_pressure,cloudcover,cloudcover_low,cloudcover_mid,cloudcover_high,et0_fao_evapotranspiration,vapor_pressure_deficit,windspeed_10m,windspeed_100m,winddirection_10m,winddirection_100m,windgusts_10m,soil_temperature_0_to_7cm,soil_temperature_7_to_28cm,soil_temperature_28_to_100cm,soil_temperature_100_to_255cm,soil_moisture_0_to_7cm,soil_moisture_7_to_28cm,soil_moisture_28_to_100cm,soil_moisture_100_to_255cm&apikey={api_key}"

        try:
            response = requests.get(url, timeout=10)
            response.raise_for_status() 
            json_data = response.json()
            save_response_to_file(response=json_data, file_name=file_name, save_dir=save_dir)

            # print(json_data)
            
        except requests.RequestException as e:
            print(f"Error for index {row['ID']}, {latitude}, {longitude}, {date}: {str(e)}")
    
    return

# df.to_csv('./merged_df_v2.csv')
```
